```python
import cv2
from matplotlib import pyplot as plt
```

```python
PATH_TO_IMAGE = '/home/jasper/dev/robocup_project/reference-images/full/converted-two-robots-2.png'
```

```python
%matplotlib inline
```

```python
img = cv2.imread(PATH_TO_IMAGE)
clahe = cv2.createCLAHE(clipLimit=6.0,
                        tileGridSize=(8, 8))
processed = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
processed[:, :, 0] = clahe.apply(processed[:, :, 0])
bgr = cv2.cvtColor(processed, cv2.COLOR_YUV2BGR)
```

```python
plt.figure(figsize=(12, 8))
plt.imshow(cv2.blur(bgr, (3, 3)))
```
